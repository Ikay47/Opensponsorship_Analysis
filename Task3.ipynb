```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dt = pd.read_csv('Athlete Search Test.csv')
ats = pd.read_csv('athlete_sports.csv')
print(dt)
print(ats)
```

### Join the two tables where the second table athletes_sports was created using unique athlete's name to find their respective sport

```python
df = pd.merge(dt,ats,on='athlete_name_y',how='left')
df
```

```python
df.isnull().sum()
```

```python
#Data cleaning
df['date'] = pd.to_datetime(df['date'])
df['view_created'] = pd.to_datetime(df['view_created'])
df['minPrice'].fillna(0,inplace= True)
df['maxPrice'].fillna(0,inplace= True)
df.isnull().sum()
```

## Task Solutions

### What trends do you see for plan level 0, plan level 3 and plan level 9? (ex. plan level 3 has a high volume of search for athletes that play basketball)


```python
#Filter all non-blanks in sport column
spt = df[df['Sport'].notnull()]
spt
```

```python
#Data grouped by planlevel
grouped = spt.groupby(['planLevel', 'Sport'])['search_id'].count().unstack()
```

```python
#Only plans 0 and 3 were found in the provided dataset
grouped
```

```python
#The grouped table is transposed
grp = grouped.T
grp
```

```python
planx = grp.sort_values(by= 0,ascending = False)
planx
```

```python
plan_0 = planx[0].head(10)
plan_0
```

```python
plany = grp.sort_values(by= 3,ascending = False)
plany
```

```python
plan_3 = plany[3].head(10)
plan_3
```

```python
print(plan_0)
print(plan_3)
```
