## Task solutions

### What is the ratio of brands that published campaigns vs. did not publish any campaign within each plan level?


```python
# Convert 'published_dates' column to a boolean column 'published'
dt['published_status'] = dt['published'].notnull()

# Convert True/False values to 'True'/'False' strings in the new column
dt['published_status'] = dt['published_status'].replace({True: 'True', False: 'False'})
dt
```

```python
# Group the DataFrame by 'Plan_Level' and calculate the count of brands with published and not published campaigns
grouped = dt.groupby(['planLevel', 'published_status'])['campaign_id'].count().unstack()
```

```python
grouped.fillna(0,inplace= True)
grouped['False']=grouped['False'].astype(int)
grouped['True']=grouped['True'].astype(int)
grouped
```

```python
grouped['ratio'] = grouped['True'].astype(str) + ':' + grouped['False'].astype(str)
```

```python
grouped
```
