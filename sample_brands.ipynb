```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Remove blanks in the following columns
dt.dropna(subset=['lastActive'],inplace=True)
dt.dropna(subset=['role'],inplace=True)
dt.dropna(subset=['industry_categories'],inplace=True)
dt.dropna(subset=['onboarding_stage'],inplace=True)
```

```python
dt.isnull().sum()
```

```python
dt.shape
```

```python
#Remove this cell as it is not correct 
dt = dt[dt['onboarding_stage'] != 12]
```

```python
dt.shape
```

```python
#Fill in the blanks
dt['bantScore'].fillna(0,inplace= True)
dt['companySize'].fillna('Not Indicated',inplace= True)
dt['country'].fillna('Not Indicated',inplace= True)
dt.isnull().sum()
```

```python
dt.shape
```

```python
dt['industry_categories']
```

max_length = dt['industry_categories'].apply(len).max()
new_columns = [f'column_{i+1}' for i in range(max_length)]

```python
data_types = dt.dtypes
print(data_types)
```

```python
# Convert the 'created' and 'lastActive' columns to datetime 
dt['created'] = pd.to_datetime(dt['created'])
dt['lastActive'] = pd.to_datetime(dt['lastActive'])
```

```python
# Calculate the date difference between 'created' and 'lastActive'
dt['active_duration'] = dt['lastActive'] - dt['created']

print(dt)
```

```python
# Set 'created' as the DataFrame index
dt.set_index('created', inplace=True)

# Resample the data to daily frequency and count the number of brand accounts created each day
dt_counts = dt.resample('M').size()

# Plotting the line chart
plt.figure(figsize=(15, 9))
plt.plot(dt_counts.index, dt_counts.values, marker='o', linestyle='-')
plt.xlabel('Date')
plt.ylabel('Brand Account Creation Count')
plt.title('Account Creation Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(dt['lastActive'], bins=20, edgecolor='black')
plt.xlabel('Last Active Date')
plt.ylabel('Frequency')
plt.title('Account Activity - Last Active Dates')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Count the occurrences of each subscription plan
plan_counts = dt['planLevel'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(plan_counts, labels=plan_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Subscription Plan Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.
plt.legend(title='Plan Level', labels=['Free', 'Trial', 'Search', 'Campaign', 'Full-Service Agency'])
plt.show()
plan_counts
```
